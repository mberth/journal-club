## Main results

- A 2D map of biomedical literature, using PubMedBERT and t-SNE on 21 million PubMed abstracts,
- Applications:
    + the rise of Covid-19 literature,
    + neuroscience's progression,
    + machine learning's influence,
    + gender disparities in authorship
    + retracted papers,
- An interactive web tool for further exploration.

## Methods

Paper abstracts (n = 21 M) --(PubMedBERT)->  768-dimensional vectors --(t-SNE)--> 2D points

## t-SNE (t-distributed Stochastic Neighbor Embedding)

1. **Dimensionality Reduction**:
    768-dimensional vectors ---> 2D points
3. **Preserving Local Structures**: "Affinities"
4. **Cost Function & Optimization**: Minimize the Kullback-Leibler divergence between distributions

## Take home

- Interactive manipulation of millions of data points in the browser is feasible
- t-SNE is useful for viualizing high-dimensional data
- Needs expertise to apply it well on larg edatasets

Run times:

> Parsing the XML files took 10 hours, computing the PubMedBERT embeddings took 74 hours, running t-SNE took 8 hours.
